```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
```

### 데이터 불러오기

```python
train = pd.read_csv('house/train.csv', index_col='Id')
print(train.shape)
train.head(2)
```

```python
test = pd.read_csv('house/test.csv', index_col='Id')
print(test.shape)
test.head(2)
```

```python
set(train.columns) - set(test.columns)
```

```python
df = pd.concat([train, test])
df.shape
```

### target

```python
sns.displot(data=train, x='SalePrice', aspect=3, rug=True, kind='kde')
plt.show()
```

```python
train["SalePrice_log1p"] = np.log1p(train["SalePrice"])
fig, axes = plt.subplots(nrows = 1, ncols=2, figsize=(12, 4))
sns.kdeplot(train["SalePrice"], ax=axes[0], fill=True)
sns.kdeplot(train["SalePrice_log1p"], ax=axes[1], fill=True)
plt.show()
```

### 결측치

```python
# 결측치 확인
isna_sum = df.isnull().sum()
isna_mean = df.isnull().mean()

pd.concat([isna_sum, isna_mean], axis=1).nlargest(10, 1)
```

```python
# 결측치가 너무 많은 값은 제거
df = df.drop(columns = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu'])
```
